```python
def get_color(q, hue):
    key = (q, hue)
    if key in cols:
        # print(f"Key {key} found in cols dictionary!")
        return tuple(cols[key][:4])  # return the RGB values as a tuple
    else:
        # print(f"Key {key} not found in cols dictionary!")
        return (0.5, 0.5, 0.5, 0.5)  # or some other default color
    
# [get_color(q, p) for q, p in zip(MI_idx.question, MI_idx.politics)]
```

```python
palette_dict = {q: {hue: get_color(q, hue) for hue in MI_idx.politics.unique()} for q in MI_idx.question.unique()}
        
for q, pol in zip(MI_idx.question, MI_idx.politics):
    color = palette_dict[q][pol]

# color
```

```python
flattened_palette = {}
for condition, palette_dict in palette_dict.items():
    for politics, color in palette_dict.items():
        key = f"{politics}_{condition}"
        flattened_palette[key] = color

flattened_palette        
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D

def custom_barplot(x, y, hue, data, width, palette, **kwargs):
    x_categories = data[x].unique()
    hue_categories = data[hue].unique()
    
    for i, x_val in enumerate(x_categories):
        x_data = data[data[x] == x_val]
        for j, hue_val in enumerate(hue_categories):
            hue_data = x_data[x_data[hue] == hue_val]
            kwargs_no_color = {k: v for k, v in kwargs.items() if k != "color"}
            edge_color = "blue" if "liberal" in hue_val else "red"
            plt.bar(i - width / 2 + j * width / len(hue_categories),
                    hue_data[y].mean(),
                    width=width / len(hue_categories),
                    color=palette[hue_val],
                    edgecolor=edge_color,
                    linewidth=1.5,
                    **kwargs_no_color)

# Create FacetGrid with the custom barplot function
g = sns.FacetGrid(data=MI_idx, col="condition", col_wrap=5, sharex=False, sharey=True, height=3, aspect=1)
g.map_dataframe(custom_barplot, x="question", y="mean", hue="hue", width=13, palette=flattened_palette)

# Create a legend outside the FacetGrid
legend_elements = [Line2D([0], [0], color=color, lw=4, label=label) for label, color in flattened_palette.items()]
plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1.9), loc='upper left', title='Hue', title_fontsize=12)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.lines import Line2D

def custom_barplot(x, y, hue, data, width, palette, **kwargs):
    x_categories = data[x].unique()
    hue_categories = data[hue].unique()
    
    for i, x_val in enumerate(x_categories):
        x_data = data[data[x] == x_val]
        for j, hue_val in enumerate(hue_categories):
            hue_data = x_data[x_data[hue] == hue_val]
            kwargs_no_color = {k: v for k, v in kwargs.items() if k != "color"}
            edge_color = "blue" if "liberal" in hue_val else "red"
            plt.bar(i - width / 2 + j * width / len(hue_categories),
                    hue_data[y].mean(),
                    width=width / len(hue_categories),
                    color=palette[hue_val],
                    edgecolor=edge_color,
                    linewidth=1.5,
                    **kwargs_no_color)
    
    # Set x-axis limits for each facet
    g.set(xlim=(-8, 14))

    # Set x-tick labels to abbreviations of each question
    abbreviations = [q[:3] for q in x_categories]
    # tick_positions = [i for i in range(len(x_categories))]
    tick_positions = [-6, -3.5, -1, 1.8, 4.3, 7, 9.5, 12.3]
    plt.xticks(tick_positions, abbreviations, ha="center")

# Create FacetGrid with the custom barplot function
g = sns.FacetGrid(data=MI_idx, col="condition", col_wrap=5, sharex=False, sharey=True, height=3, aspect=1)
g.map_dataframe(custom_barplot, x="question", y="mean", hue="hue", width=13, palette=flattened_palette)

# Title
g.set_titles(fontweight='bold', col_template="{col_name}")

# flattened_palette is assumed to be a dictionary mapping label names to colors
unique_colors = list(set(flattened_palette.values()))

# Create a dictionary mapping colors to labels
# color_labels = {color: f"My label {i}" for i, color in enumerate(unique_colors)}

# Specify custom labels for each color
custom_labels = ["punishment", "genetics", "Label 3"] # replace with your custom labels

# Create legend elements for each unique color
# legend_elements = [Line2D([0], [0], color=color, lw=4, label=color_labels[color]) for color in unique_colors]
legend_elements = [Line2D([0], [0], color=unique_colors[i], lw=4, label=custom_labels[i]) for i in range(len(unique_colors))]

# Create a legend outside the FacetGrid
plt.legend(handles=legend_elements, bbox_to_anchor=(1.05, 1.9), loc='upper left', title='Hue', title_fontsize=12)

plt.show()
```
