## MFW

When people see things as morally wrong, there is a positive association between perceived mental illness and increased punitiveness. When things are not seen as morally wrong, there is a negative association between between perceived mental illness and punitiveness. Potential explanation: mental illness exonerates people unless people are highly motivated to morally judge the condition, in which case mental illness is viewed as another thing wrong with the person that makes punishment appropriate.

```python
for i in range(0, 10):
    for j in MI_conditions.iloc[:, 9*i:9*i+9]:
        for a in MI_conditions.iloc[:, 9*i:9*i+9]:
            if j != a:
                J = MI_conditions[j].dropna()
                A = MI_conditions[a].dropna()
                m, b = np.polyfit(J, A, 1)
                plt.hist2d(J, A, cmap= 'Blues')
                plt.plot(J, m*J + b)
                
```
