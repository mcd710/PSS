```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.linalg import svd
from sklearn import linear_model
import pandas as pd
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
import matplotlib.cm as cm
from IPython.display import display
from mpl_toolkits.mplot3d import Axes3D
from sklearn.feature_selection import f_regression
from statsmodels.stats.anova import anova_lm

from sklearn.preprocessing import PolynomialFeatures
import scipy.special
```

```python
MI = pd.read_csv("/Users/merieldoyle/Desktop/CK/MI_cleaned.csv", index_col = [0])
MI = pd.DataFrame(MI)
MI_conditions = MI.drop(MI.iloc[:, 90:153], axis = 1)

MI_conditions
```

```python
#recall that some values need to be dropped for the s and e politics columns 
np.where(MI["social_politics"] == -99), np.where(MI["economic_politics"]== -99)
drop_pols = [66, 172, 578, 618, 893, 976, 988, 1102] # rows that need to be dropped due to ^condition
```

```python
for i in np.arange(0, 10): #for each of the conditions
    subset = MI_conditions.iloc[:, 9*i:9*i + 9] #subset made for a single condition
    dgs = MI[["ladder", "sex", "age", "marital_stat", "education", "politics", "social_politics", "economic_politics", "religiosity"]]
    for j in drop_pols:
        subset = subset.drop(j) #dropping the values that are -99 in pols
    subset = subset.join(dgs).dropna(axis = 0) #with demographics and dropping NA
    subset_y = subset.iloc[:, 4:5].reset_index(drop = True) #using punishment as the y
    subset_x = subset.drop(subset.iloc[:, 4:5], axis = 1) 
    
    subcol = subset.columns
    poly = PolynomialFeatures(interaction_only=True, degree = 2)
    Xtr = pd.DataFrame(data = poly.fit_transform(subset_x), columns = poly.get_feature_names(subset_x.columns))
    subset_x = subset_x.reset_index(drop = True)
    Xt = subset_x.join(Xtr[subcol[0] + ' ' +subcol[6]]) 
    
    #Mulitvariate regresssion 
    Xt = sm.add_constant(Xt)
    mreg = sm.OLS(subset_y, Xt).fit()
    display(mreg.summary())
```
