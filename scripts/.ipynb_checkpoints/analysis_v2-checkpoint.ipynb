## Meriel - Cory Clark - Motivated Free Will Dataset

### Comments from Clark: When people see things as morally wrong, there is a positive association between perceived mental illness and increased punitiveness. When things are not seen as morally wrong, there is a negative association between between perceived mental illness and punitiveness. Potential explanation: mental illness exonerates people unless people are highly motivated to morally judge the condition, in which case mental illness is viewed as another thing wrong with the person that makes punishment appropriate.

```python
MI_condnew = MI_conditions
MI_condnew = MI_condnew.drop(["pedophile_common", "drugs_common", "psycho_common", "gay_common", "fetish_common", "racist_common", "trans_common", "obese_common", "dep_common", "schizo_common"], axis = 1)
    
MI_condnew;
```

```python
w_pc = []
error_cond = []
for i in np.arange(0, 10): #for each of the conditions
    subset = MI_condnew.iloc[:, 8*i:8*i + 8] #subset made for a single condition
    subset_y = subset.iloc[:, 6:7] #using wrongness as the y
    subset_x = subset.drop(subset.iloc[:, 6:7], axis = 1) #using the rest as the x (predictors)

    #Cross validation part
    for j in np.arange(0, 13): #for each of the 13 sections of the data (94 pt in ea)
        subset_x_test = subset_x.iloc[94*j:94*j + 94, :] #selecting 1/13 as hold out
        subset_y_test = subset_y.iloc[94*j:94*j + 94, :] #corresponding y
        subset_x_train = subset_x.drop(subset_x.index[94*j:94*j + 94], axis = 0) #taking out 1/13 for train 
        subset_y_train = subset_y.drop(subset_y.index[94*j:94*j + 94], axis = 0) #corresponding y
        
        subset_x_train.dropna(inplace = True) #dropping NA values so we can do SVD
        subset_y_train.dropna(inplace = True)
        subset_x_test.dropna(inplace = True)
        subset_y_test.dropna(inplace = True) 
        
        U, S, V = np.linalg.svd(subset_x_train) #find w using SVD, pseudo inverse
        S = np.multiply(1/S, np.identity(V.shape[0]))
        right_dim = np.zeros([7, (U.shape[0]-7)])
        S = np.concatenate((S, right_dim), 1)
        w = np.dot(V, S).dot(np.transpose(U)).dot(subset_y_train)
        
        w_pc = np.append(w_pc, w) 
        
        error = []
        y_hat = np.round(np.dot(subset_x_test, w), decimals = 0) #use that to predict the y 
        error = np.mean(y_hat != subset_y_test)
    error_cond = np.append(error_cond, error)
    
print(error_cond)
```
